```python
# Loading housing-covid data into dataframe and converting the date column to datetime
housing_covid_df = pd.read_csv("../data/processed data/housing_covid_final.csv")
housing_covid_df['Date'] = pd.to_datetime(housing_covid_df['Date'])
print(housing_covid_df.dtypes, housing_covid_df.shape)
housing_covid_df.head(3)
```

```python
# CREATE FUNCTION FOR CUSTOM LINE PLOT
white_grid = sns.set_style("whitegrid")

# Create an array with the colors you want to use
color_CA = ["#EB8484","#9E0508","#F8171B","#551011","#FFC0CB","#FF3E96","#A46582","#584E56","#660198","#3300FF","#9999FF","#457CBE","#52D5F5","#4DFFFF","#4DFFDB","#3E7A5E","#00FE7E","#F4BF3E","#FF5B09"]
# color list - black,gray,red,maroon,green,royal blue,purple,dark green,lavender, light brown,orange
color_general = ["#131313","#6d727a","#FF0000","#821916","#00FF00","#4169e1","#731877","#158b64","#dcb3fd","#c09a85","#FFA500"]
    
def line_plot(colors,x,y,hue,data,title,xlabel,ylabel):
    # Set your custom color palette
    sns.set_palette(sns.color_palette(colors))
    plt.figure(figsize=(15, 5))
    sns.lineplot(x=x, y=y, hue=hue, data=data)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(rotation=45)
    plt.legend(bbox_to_anchor=(1.2, 1), loc='upper right', borderaxespad=0)
    plt.show() 
```

The housing prices between 2000 to 2021 for most of the states, shows a dip around late 2008 to early 2012. This is because of the Great Recession that began in December 2007 and ended in June 2009. This maybe beacuse commuting reduced and people wanted to buy houses on the outskirts, instead of paying for house rents in the metro areas.

# 30-Year Fixed Rate Mortgage Average in the United States: Data cleaning

We can see if the average mortgage interest rates in USA has any impact on the housing prices. The data is extracted from: https://fred.stlouisfed.org/series/MORTGAGE30US . There is no state wise or county wise interest rate data. The data is captured on thursday every week. We can aggregate monthly average interest rate data and map with all the state-counties monthly data. 

```python
# Loading 30-Year Fixed Rate Mortgage Average in the United States data into dataframe and converting the date column to datetime
mortgage_df = pd.read_csv("../data/raw data/MORTGAGE30US.csv")
mortgage_df['DATE'] = pd.to_datetime(mortgage_df['DATE'])
mortgage_df.head(3)
```

```python
mortgage_df.set_index('DATE', inplace=True)
mortgage_df.info()
```

```python
mortgage_df = mortgage_df.resample('M').mean()
```

```python
mortgage_df.tail()
```

```python
mortgage_df.reset_index(inplace=True)
mortgage_df.head()
```

```python
mortgage_df.dtypes
```

```python
mortgage_df = mortgage_df[(mortgage_df['DATE'] > '1999-12-31')]
mortgage_df.reset_index(inplace=True)
mortgage_df.drop(columns="index", inplace=True)
print(mortgage_df.shape)
mortgage_df.head()
```

```python
housing_full_df.columns
```

```python
# Joining housing price and mortgage interest rate data
housing_int_df = pd.merge(housing_full_df, mortgage_df,  how='inner', left_on=['Date'], right_on = ['DATE'])
print(housing_int_df.shape)
housing_int_df.tail()
```

```python
housing_int_df.drop(columns="DATE", inplace= True)
housing_int_df.rename(columns= {"MORTGAGE30US" : "mortgage_int_rate"}, inplace= True)
housing_int_df.head()
```

```python
housing_int_df.to_csv("../data/processed data/housing_mortgage_final.csv", index = False)
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(housing_int_df.corr(method='pearson'), annot = True);
```

<b>Looks like there is a small negative correlation between house price and mortgage interest rate.</b>
